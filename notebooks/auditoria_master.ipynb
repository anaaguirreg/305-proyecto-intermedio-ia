```python
import pandas as pd
from pathlib import Path

PROJECT_ROOT = Path.cwd() if (Path.cwd() / "config").exists() else Path.cwd().parent
master_path = PROJECT_ROOT / "data/master" / "master_table.parquet"

df = pd.read_parquet(master_path)

# Tomar las 3 filas de un municipio-año cualquiera
sample = df.groupby(['cod_municipio', 'anio_hecho']).head(3)
print("🔍 Columnas que DIFERENCIAN las 3 filas (donde hay variación):")
diff_cols = []
for col in sample.columns:
    if sample[col].nunique() > 1:
        print(f"   ✅ '{col}' varía → valores: {sample[col].unique().tolist()}")
        diff_cols.append(col)

if not diff_cols:
    print("   ⚠️ NO hay columnas distintas. Son DUPLICADOS PUROS (misma información repetida 3 veces).")
else:
    print(f"\n📌 Causa raíz identificada: {diff_cols}")
```
